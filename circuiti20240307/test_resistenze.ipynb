## STUDIO DELL'INCERTEZZA SUL VALORE DELLE RESISTENZE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

dir_path = os.path.abspath('')
sys.path.append(dir_path + '/../')

from interpolazione import Interpolazione
```

```python
dir_path = os.path.abspath('')
FILE = dir_path + '/data/' + 'test_resistenze.csv'
fr = pd.read_csv(FILE)
Rattese = np.array(fr['R attesa']) #Î©
exponents = np.array(fr['exponent']) 
Rattese = Rattese * 10**exponents

R = np.array(fr['Rmisurata'])
R = R* 10**exponents

R = R - 0.2

N = np.arange(0,len(R))
```

```python
plt.plot(N,R,linestyle='None',marker='.')
plt.plot(N,Rattese,linestyle='None',marker='.')
plt.yscale('log')
plt.show()
```

```python
N = len(R)
err = np.sum(np.abs((Rattese - R)/R))/N
print(err)

err = np.sum(np.abs((Rattese - R)/Rattese))/N
print(err)
```
